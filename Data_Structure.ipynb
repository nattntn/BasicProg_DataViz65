## EXAMPLE คำนวนเวลาเป็นวินาทีของเวลาต่อไปนี้โดยใช้คำสั่ง split() ช่วย (print ออกมาให้สวยงาม)
- 12:30:15
- 13:41:07
- 09:11:50
- 21:30:15
- 03:10:15

````python
# 01:02:03

a = 1*60*60 + (2*60) + 3
print(f'01:02:03 แปลงเป็นวินาที่ได้ {a} วินาที')
````

````python
# 12:30:15
B = (12*60*60) + (30*60) + 15
print(f'12:30:15 แปลงเป็นวินาที ได้ {B} วินาที')
````

````python
ex_B = '12:30:15'
print(ex_B)
````

````python
sp_ex_B = ex_B.split(':')
print(sp_ex_B)
````

````python
total_sec = (int(sp_ex_B[0])*60*60) + (int(sp_ex_B[1])*60) + int(sp_ex_B[2])
print(total_sec)
````

````python
print(f'เวลา {ex_B} สามารถแปลงเป็นหน่วยวินาทีได้ {total_sec:,} วินาที')
````

````python
C = (13*60*60) + (41*60) + 7
print(f'13:41:07 แปลงเป็นวินาที ได้ {C} วินาที')
````

````python
D = (9*60*60) + (11*60) + 50
print(f'09:11:50 แปลงเป็นวินาที ได้ {D} วินาที')
````

````python
E = (21*60*60) + (30*60) + 15
print(f'21:30:15 แปลงเป็นวินาที ได้ {E} วินาที')
````

````python
F = (3*60*60) + (10*60) + 15
print(f'03:10:15 แปลงเป็นวินาที ได้ {F} วินาที')
````

## 1. หาพื้นที่ (district_name) ที่มีจำนวนโครงการเยอะที่สุดและมีความต้องการเช่าเยอะที่สุด (```5 คะแนน```)

````python
rent.groupby('listing_district_name_en')[['listing_district_name_en','project_propertytype_name_en','volume_of_rental']].max() .max()
````

## 2. โครงการที่แพงที่สุดในพื้นที่ที่มีความต้องการเช่าสูงที่สุด (```5 คะแนน```)

````python
unit.merge(project, left_on='project_id', right_on='project_id')
````

````python
project_nodup = project.drop_duplicates(subset=['project_id'])
````

````python
unit['pricemin'] = unit['project_id'].map(project_nodup.set_index('project_id')['price_min'])
````

````python
unit
````

## 3. ราคาเฉลี่ยของโครงการ ที่อยู่ในย่าน (subdistrict) ที่  อาหารการกินมีความหลากหลายที่สุดในช่วงกลางวัน (eating_daytime_variety) ของประเทศไทย (```5 คะแนน```)

## 5. (เฉพาะ กรุงเทพมหานคร) เปรียบเทียบราคาโครงการที่แพงที่สุดในย่าน (subdistrict)ที่ราคาของอาหารการกินในช่วงกลางคืนที่ราคาสูงที่สุด (eating_nighttime_price) กับ ย่าน (subdistrict)ที่ราคาของอาหารการกินในช่วงกลางคืนที่ราคาต่ำที่สุด (eating_nighttime_price) (```10 คะแนน```)

````python
eat_bkk = eat[eat['province_name_en']=='Bangkok']
eat_bkk
````

````python
eat_bkk_nighttime = eat_bkk.groupby('subdistrict_name_th')['eating_nighttime_price'].max()
eat_bkk_nighttime
````

````python
eat_bkk_daytime = eat_bkk.groupby('subdistrict_name_th')['eating_daytime_price'].min()
eat_bkk_daytime
````

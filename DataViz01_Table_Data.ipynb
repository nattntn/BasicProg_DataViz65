# Parallel Coordinates

```python
pd.plotting.parallel_coordinates(df, 'Name') #python  pd.plotting.parallel_coordinates(จากตาราง, 'จัดกลุ่มโดย')
# จากกราฟ เขียวสด กับม่วง คล้ายกัย ส่วนเขียวอ่อน จะเล็กต่างจากเพื่อน
```

### select columns to show

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength','SepalWidth','PetalWidth']) 
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']) # cols = ['SepalLength','PetalLength'] เลือกเฉพาะ คอลัมน์ที่อยากทำมาเเสดง
#กลีบดอกกับกลีบเลี้ยง มีขนาดต่างกันมั้ย
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']);
```

## Subplot

```python
matplotlib.rcParams['figure.figsize']=[10,10]
```

```python
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,3,3)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest',cmap='hot')
```

```python
plt.subplot(1,4,1)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,4,2)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.2)

plt.subplot(1,4,3)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.5)

plt.subplot(1,4,4)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.8)
```

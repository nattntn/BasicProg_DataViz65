# Parallel Coordinates

```python
pd.plotting.parallel_coordinates(df, 'Name') #python  pd.plotting.parallel_coordinates(จากตาราง, 'จัดกลุ่มโดย')
# จากกราฟ เขียวสด กับม่วง คล้ายกัย ส่วนเขียวอ่อน จะเล็กต่างจากเพื่อน
```

### select columns to show

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength','SepalWidth','PetalWidth']) 
```

```python
pd.plotting.parallel_coordinates(df, 'Name',cols = ['SepalLength','PetalLength']) # cols = ['SepalLength','PetalLength'] เลือกเฉพาะ คอลัมน์ที่อยากทำมาเเสดง
#กลีบดอกกับกลีบเลี้ยง มีขนาดต่างกันมั้ย
```

# Matplotlib

```python
from matplotlib import pyplot as plt #เพื่อ plot graph
```

```python
import matplotlib #เพื่อปรับแต่งขนาดของกราฟ
```

## Heatmap matrix

```python
matplotlib.rcParams['figure.figsize']=[30,30] #กำหนดขนาดของ figure
plt.imshow(df.iloc[:,:-1],interpolation='nearest') # -1 ไม่เอาคอลัมน์สุดท้าย  // เข้ม มาก  สว่างน้อย
```

```python
plt.imshow(df.iloc[:,:-1],interpolation='nearest',cmap = 'hot')
```

## Subplot  // พอตหลายกราฟมาวางข้างกัน

```python
matplotlib.rcParams['figure.figsize']=[20,20] # กำหนดขนาดกราฟ
```

```python
plt.subplot(1,3,1)  # subplot(มีกี่แถว, มีกี่หลัก, เป็นตัวที่เท่าไหร่)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest') # พอต 50 ตัวแรก

plt.subplot(1,3,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest')

plt.subplot(1,3,3)
plt.imshow(df.iloc[100:150,:-1],interpolation='nearest')
```

```python
plt.subplot(1,4,2)
plt.imshow(df.iloc[50:100,:-1],interpolation='nearest',alpha = 0.8)

plt.subplot(1,4,1)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot')

plt.subplot(1,4,2)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.6) #alpha =  ความโปร่งใสของขนาด figure

plt.subplot(1,4,3)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.5)

plt.subplot(1,4,4)
plt.imshow(df.iloc[:50,:-1],interpolation='nearest',cmap='hot',alpha = 0.8)

# alpha ทำให้มองเห็นจุดที่ถูกทับ
```

```python
# RGBA (Alpha)
plt.subplot(1,2,1)
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0.5,0.5,0.5],[0,0,0.8,0.5],[0.2,0.4,0.7,0.5]]) # [0.5,0.5,0.5,0.5 (alpha)]

plt.subplot(1,2,2)
pd.plotting.parallel_coordinates(df, 'Name', color=[[0.5,0.5,0.5],[0,0,0.8],[0.2,0.4,0.7]])
```

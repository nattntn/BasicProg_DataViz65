## ลบ Missing

missing ->  null, na(not autorized, NaN (not a number)

```python
unit.dropna()  # axis =0 พิจารณาแนวแถว/ 1 พิจารณา ตามหลัก 
```

```python
unit.dropna(axis=1) # เอาเเค่คอลัมที่ไม่มีmissing เลย
```

how = 'any' / มี 1  missing ก็จะโดนตัดเลย เเต่ถ้า = 'all' ก็ต้องเป็นทุกอันที่มีค่าว่าง ถึงตัด

```python
unit.dropna(how = 'all')
```

thresh : int ใส่จำนวนที่มีค่า missing เกินกี่ค่า ถึงจะลบ

```python
unit.iloc[0,:]
```

```python
unit.dropna(axis=1,thresh=42000)
```

```python
.max()= หา มากสุดของเเต่ละคอลัมน์
```

อาจารย์เฉลย

```python
unit.groupby('propertytype_name_en')[['price_min','area_total_min_wa','area_usable_min','count_room_bed']].max() 
```

```python
maxprice = unit.groupby('propertytype_name_en')[['price_min','area_total_min_wa','area_usable_min','count_room_bed']].max() .reset_index()
maxprice
```

```python
for row in maxprice.iterrows():
    print(row[1]['propertytype_name_en'],row[1]['price_min'])
    print(unit[(unit['propertytype_name_en']==row[1]['propertytype_name_en']) & (unit['price_min']==row[1]['price_min'])][['area_total_min_wa','area_usable_min','count_room_bed']])
```

```python
unit[(unit['propertytype_name_en']==row[1]['propertytype_name_en']) & (unit['price_min']==row[1]['price_min'])][['area_total_min_wa','area_usable_min','count_room_bed']]
```

3.จำนวนที่จอดรถและห้องน้ำของทรัพย์สินที่ถูกที่สุดแต่ละชนิด

```python
unit.groupby('propertytype_name_en')[['price_min','count_parking','count_room_bath']].min() 
```

```python
minprice = unit.groupby('propertytype_name_en')[['price_min','count_parking','count_room_bath']].min() .reset_index()
minprice
```

```python
for row in minprice.iterrows():
    print(row[1]['propertytype_name_en'],row[1]['price_min'])
    print(unit[(unit['propertytype_name_en']==row[1]['propertytype_name_en']) & (unit['price_min']==row[1]['price_min'])][['count_parking','count_room_bath']])
```
